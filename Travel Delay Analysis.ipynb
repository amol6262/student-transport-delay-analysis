```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("student_transport_delay_extended.csv")

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

required_cols = ["transport", "delay_minutes"]

for col in required_cols:
    if col not in df.columns:
        print(f"CSV must contain '{col}' column")
        exit()


# Mean & Standard Deviation

mean_delay = df.groupby("transport")["delay_minutes"].mean()
std_delay = df.groupby("transport")["delay_minutes"].std()

print("STUDENT TRANSPORT DELAY REPORT (EXTENDED DATASET)\n")
print("Total Records:", len(df))

print("\nAverage Delay (Minutes):")
print(mean_delay)

print("\nStandard Deviation (Minutes):")
print(std_delay)


#  Bar Graph (Average Delay)

plt.figure()
mean_delay.plot(kind="bar")
plt.title("Average Transport Delay")
plt.xlabel("Transport Type")
plt.ylabel("Average Delay (Minutes)")
plt.show()


#  Histogram (Delay Distribution)

plt.figure()
plt.hist(df["delay_minutes"], bins=15)
plt.title("Delay Distribution of Students")
plt.xlabel("Delay Minutes")
plt.ylabel("Number of Students")
plt.show()


#  Pie Chart (Transport Usage)

transport_count = df["transport"].value_counts()

plt.figure()
plt.pie(
    transport_count,
    labels=transport_count.index,
    autopct="%1.1f%%"
)
plt.title("Transport Mode Usage")
plt.show()


#  Box Plot (Delay Comparison)

plt.figure()
df.boxplot(column="delay_minutes", by="transport")
plt.title("Box Plot of Transport Delays")
plt.suptitle("")
plt.xlabel("Transport Type")
plt.ylabel("Delay Minutes")
plt.show()




```
